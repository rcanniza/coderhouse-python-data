```python
import requests

def alerta(mensagem):
    print(f"ALERTA: {mensagem}")

def extrair_dados(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        alerta(f"Erro HTTP: {http_err}")
    except Exception as err:
        alerta(f"Erro inesperado: {err}")

url_pokemon = "https://pokeapi.co/api/v2/pokemon/ditto"
url_habilidades = "https://pokeapi.co/api/v2/ability/1"
url_tipos = "https://pokeapi.co/api/v2/type/3"

dados_pokemon = extrair_dados(url_pokemon)
dados_habilidades = extrair_dados(url_habilidades)
dados_tipos = extrair_dados(url_tipos)

print("Tabela de Pokémon:")
print(dados_pokemon)

print("\nTabela de Habilidades:")
print(dados_habilidades)

print("\nTabela de Tipos:")
print(dados_tipos)
```
